# Data Cleaning


```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option("display.max_columns", None)
```

```python
train = pd.read_csv("../data/raw/train.csv")
test = pd.read_csv("../data/raw/test.csv")

print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
target = "SalePrice"

y = train[target]
train_features = train.drop(columns=[target])

print("Train features shape:", train_features.shape)
print("Target shape:", y.shape)
```

## Combine train and test for consistent cleaning


```python
train_features["Dataset"] = "train"
test["Dataset"] = "test"

combined = pd.concat([train_features, test], axis=0, ignore_index=True)

print("Combined shape:", combined.shape)
```

## Remove ID column


```python
if "Id" in combined.columns:
    combined = combined.drop(columns=["Id"])

combined.head()
```

## Handle missing categorical values

```python
categorical_cols = combined.select_dtypes(include=["string", "object"]).columns.tolist()
categorical_cols.remove("Dataset")

print("Missing categorical values before cleaning:")
print(combined[categorical_cols].isnull().sum().sum())

for col in categorical_cols:
    combined[col] = combined[col].fillna("None")

print("Missing categorical values after cleaning:")
print(combined[categorical_cols].isnull().sum().sum())
```

```python
numerical_cols = combined.select_dtypes(include=["int64", "float64"]).columns.tolist()

print("Missing numerical values before cleaning:")
print(combined[numerical_cols].isnull().sum().sum())

for col in numerical_cols:
    combined[col] = combined[col].fillna(combined[col].median())

print("Missing numerical values after cleaning:")
print(combined[numerical_cols].isnull().sum().sum())
```

```python
total_missing = combined.isnull().sum().sum()
print("Total missing values:", total_missing)
```

## Split cleaned data back into train and test


```python
cleaned_train = combined[combined["Dataset"] == "train"].drop(columns=["Dataset"])
cleaned_test = combined[combined["Dataset"] == "test"].drop(columns=["Dataset"])

cleaned_train[target] = y.values

print("Cleaned train shape:", cleaned_train.shape)
print("Cleaned test shape:", cleaned_test.shape)
```

## Save cleaned data

```python
Path("../data/processed").mkdir(parents=True, exist_ok=True)

cleaned_train.to_csv("../data/processed/cleaned_train.csv", index=False)
cleaned_test.to_csv("../data/processed/cleaned_test.csv", index=False)

print("Saved cleaned_train.csv and cleaned_test.csv successfully.")
```
